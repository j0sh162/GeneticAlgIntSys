```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load data from CSV
data = pd.read_csv("knapsack_GA_results.csv")

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(data["Generation"], data["Best Value"], linestyle='-')
plt.title("Genetic Algorithm Performance")
plt.xlabel("Generation")
plt.ylabel("Best Value")
plt.grid(True)
plt.show()
```
